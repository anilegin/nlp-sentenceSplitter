# Evaluate Final Sentence Splitting Ensemble

In this notebook, you can evaluate the final language-specific sentence splitting pipeline starting from a **raw text file**.

Given a language and a raw input path, the notebook preprocesses the file into candidate boundaries, loads the corresponding CharCNN, BiLSTM, and meta-classifier, and produces final sentence-boundary predictions and reconstructed split text.

```python
# RUN THESE IF YOU ARE ON COLAB
!git clone https://github.com/anilegin/nlp-sentenceSplitter.git
!cd nlp-sentenceSplitter/
import sys
sys.path.append('/content/nlp-sentenceSplitter')
```

```python
from pathlib import Path
import os
from inference import evaluate_sentence_splitter
```

```python
language = "english"   # "english" or "italian"
#data you want to test
raw_file_path = "/content/nlp-sentenceSplitter/data/raw/UD_English-GUM/en_gum-ud-dev.sent_split"





if os.path.exists("/content/nlp-sentenceSplitter/models"):
  models_root = "/content/nlp-sentenceSplitter/models"
  output_dir = "/content/nlp-sentenceSplitter/results"
else:
  models_root = "./models"
  output_dir = "./results"

results = evaluate_sentence_splitter(
    language=language,
    raw_file_path=raw_file_path,
    models_root=models_root,
    output_dir=output_dir,
    batch_size=128,
    eos_token="<EOS>",
    window=50,
)

print("Language:", results["language"])
print("Has labels:", results["has_labels"])
print("Candidate predictions saved to:", results["candidate_predictions_path"])
print("Predicted split text saved to:", results["predicted_text_path"])

if results["metrics"] is not None:
    print("CharCNN Accuracy:", results["metrics"]["charcnn_accuracy"])
    print("CharCNN F1:", results["metrics"]["charcnn_f1"])
    print("BiLSTM Accuracy:", results["metrics"]["bilstm_accuracy"])
    print("BiLSTM F1:", results["metrics"]["bilstm_f1"])
    print("Ensemble Accuracy:", results["metrics"]["ensemble_accuracy"])
    print("Ensemble F1:", results["metrics"]["ensemble_f1"])
    print(results["metrics"]["ensemble_report"])
else:
    print("No usable labels found. Predictions were saved.")
```
